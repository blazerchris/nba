```python
import numpy as np
import pandas as pd
```

```python
salarydf = pd.read_csv("NBA_Salary_96to17.csv")
rankdf = pd.read_csv("seasonRanks.csv")
```

```python
teamSalaryPerYear = list()
for year in salarydf.Year.unique():
    year_df = salarydf.loc[salarydf.Year == year]
    for team in year_df.Team.unique():
        team_df = year_df.loc[year_df.Team == team]
        salary_total = team_df.Salary.sum(axis=0)
        teamSalaryPerYear.append((year,team,salary_total))
```

```python
labels = ["Year","Team","Team_Salary"]
teamdf = pd.DataFrame.from_records(teamSalaryPerYear,columns=labels)
```

```python
teamdf
```

```python
yearlyAverage = list()
for year in teamdf.Year.unique():
    average = int(teamdf.loc[teamdf.Year == year].Team_Salary.mean())
    yearlyAverage.append((year,average))
```

```python
yearlyAverageDF = pd.DataFrame.from_records(yearlyAverage,columns=["Year","Average_Team_Salary"])
```

```python
yearlyAverageDF["pct_change"] = pd.Series(yearlyAverageDF.Average_Team_Salary.pct_change(axis=0),index=yearlyAverageDF.index)
```
