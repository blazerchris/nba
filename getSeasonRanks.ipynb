```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import collections
import csv
import time
```

```python
import numpy as py
import pandas as pd
```

```python
driver = webdriver.Chrome("/Users/Chris/Documents/chromedriver.exe")
```

```python
seasonRanks = list()
for year in range(1991,1996):
    url = "http://www.basketball-reference.com/leagues/NBA_%d.html" % year
    driver.get(url)
    e_confs = driver.find_element_by_id("divs_standings_E")
    for team in e_confs.find_elements_by_xpath('.//tbody/tr'):
        try:
            tag = team.find_element_by_xpath('.//th/a').get_attribute("href")[-13:-10]
            win = int(team.find_element_by_xpath('.//td').text)
            seasonRanks.append((year,"East",tag,win))
        except:
            pass
    w_confs = driver.find_element_by_id("divs_standings_W")
    for team in w_confs.find_elements_by_xpath('.//tbody/tr'):
        try:
            tag = team.find_element_by_xpath('.//th/a').get_attribute("href")[-13:-10]
            win = int(team.find_element_by_xpath('.//td').text)
            seasonRanks.append((year,"West",tag,win))
        except:
            pass
```

```python
for year in range(2016,2018):
    url = "http://www.basketball-reference.com/leagues/NBA_%d.html" % year
    driver.get(url)
    e_confs = driver.find_element_by_id("all_confs_standings_E")
    for team in e_confs.find_elements_by_xpath('.//tbody/tr'):
        try:
            tag = team.find_element_by_xpath('.//th/a').get_attribute("href")[-13:-10]
            win = int(team.find_element_by_xpath('.//td').text)
            seasonRanks.append((year,"East",tag,win))
        except:
            pass
    w_confs = driver.find_element_by_id("all_confs_standings_W")
    for team in w_confs.find_elements_by_xpath('.//tbody/tr'):
        try:
            tag = team.find_element_by_xpath('.//th/a').get_attribute("href")[-13:-10]
            win = int(team.find_element_by_xpath('.//td').text)
            seasonRanks.append((year,"West",tag,win))
        except:
            pass
```

```python
seasonRanks
```

```python
df = pd.DataFrame.from_records(seasonRanks,columns=["Year","Div","Team","Wins"])
```

```python
df.to_csv("seasonRanks86to95.csv",index=False)
```
