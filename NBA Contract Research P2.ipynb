```python
import matplotlib.pyplot as plt
```

```python
def top5paidPie(odf,year,team):
    df = odf.loc[(odf['Team'] == team) & (df2['Year'] == year)]
    salary_list = list(df.iloc[0:5]['Salary'])
    salary_list.append(df.iloc[5:]['Salary'].sum())
    name = list(df.iloc[0:5]['Player'])
    name.append("Rest")
    colors = ["#E13F29", "#D69A80", "#D63B59", "#AE5552", "#EB8076", "#96624E"]
    plt.pie(
    salary_list,
    labels = name,
    shadow = False,
    explode = (0.15,0,0,0,0,0),
    colors = colors,
    startangle=90,
    autopct='%1.1f%%',
    )
    plt.axis('equal')
    plt.tight_layout()
    plt.suptitle('%d %s' % (int(year), team),fontsize=14, fontweight='bold')
    plt.show()
```

```python
top5paidPie(df2,2015,"ATL")
```

```python
top5paidPie(df2,2016,"ATL")
```
