```python
import numpy as np
import pandas as pd
```

```python
salarydf = pd.read_csv("NBA_Salary_96to17.csv")
rankdf = pd.read_csv("seasonRanks.csv")
```

```python
teamSalaryPerYear = list()
for year in salarydf.Year.unique():
    year_df = salarydf.loc[salarydf.Year == year]
    for team in year_df.Team.unique():
        team_df = year_df.loc[year_df.Team == team]
        salary_total = team_df.Salary.sum(axis=0)
        teamSalaryPerYear.append((year,team,salary_total))
```

```python
labels = ["Year","Team","Team_Salary"]
teamdf = pd.DataFrame.from_records(teamSalaryPerYear,columns=labels)
```

```python
combined = pd.merge(rankdf,teamdf,on=['Year','Team'])
```

```python
combined['Season_Rank'] = combined.groupby('Year')['Wins'].rank(ascending=False)
```

```python
combined['Salary_Rank'] = combined.groupby('Year')['Team_Salary'].rank(ascending=False)
```

```python
combined['Div_Season_Rank'] = combined.groupby(['Year','Div'])['Wins'].rank(ascending=False)
```

```python
combined['Div_Salary_Rank'] = combined.groupby(['Year','Div'])['Team_Salary'].rank(ascending=False)
```

```python
combined['Diff'] = combined['Salary_Rank'].sub(combined['Season_Rank'],axis=0)
```

```python
combined
```

```python
combined.to_csv('combined.csv',index=False)
```

```python
combined['Div_Diff'] = combined['Div_Salary_Rank'].sub(combined['Div_Season_Rank'],axis=0)
```

```python
combined.to_csv("combined.csv")
```

```python
yearlyAverage = list()
for year in teamdf.Year.unique():
    average = int(teamdf.loc[teamdf.Year == year].Team_Salary.mean())
    yearlyAverage.append((year,average))
```
