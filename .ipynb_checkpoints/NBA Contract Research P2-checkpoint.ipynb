```python
def top5paid(odf,year,team):
    df = odf.loc[(odf['Team'] == team) & (odf['Year'] == year)]
    salary_list = list(df.iloc[0:5]['Salary'])
    salary_list.append(df.iloc[5:]['Salary'].sum())
    name = list(df.iloc[0:5]['Player'])
    name.append("Rest")
    return salary_list,name
```
